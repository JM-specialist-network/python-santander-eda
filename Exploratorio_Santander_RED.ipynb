```python
from scipy.stats import skew, kurtosis

# Función para calcular estadísticas descriptivas
def descriptive_stats(column):
    stats = {
        "Media": column.mean(),
        "Mediana": column.median(),
        "Desviación Estándar": column.std(),
        "Rango": column.max() - column.min(),
        "Rango Intercuartílico": column.quantile(0.75) - column.quantile(0.25),
        "Coeficiente de Asimetría": skew(column.dropna()),
        "Curtosis": kurtosis(column.dropna())
    }
    return stats

# Calcular estadísticas para 'age' y 'renta'
age_stats = descriptive_stats(df_santander['age'])
renta_stats = descriptive_stats(df_santander['renta'])

# Mostrar resultados
print("\nEstadísticas para 'age':")
print(age_stats)
print("\nEstadísticas para 'renta':")
print(renta_stats)
```

```python
# Detección de Outliers. Para ello, calcular los percentiles 0.05 (Q1) y 0.95(Q3) utilizando la función quantile() de Pandas. Luego, calcular el (pseudo)rango intercuartílico (IQR) como la diferencia entre Q3 y Q1. Luego definir los límites inferior y superior para identificar outliers como 1.5 veces el IQR por debajo de Q1 y por encima de Q3. Finalmente, encontrar los outliers al seleccionar los valores que están por debajo del límite inferior o por encima del límite superior.
```

```python
# Detectar outliers
def detect_outliers(column):
    Q1 = column.quantile(0.05)
    Q3 = column.quantile(0.95)
    IQR = Q3 - Q1 #Pseudo rango intercuartílico
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers

# Detectar outliers para 'age' y 'renta'
outliers_age = detect_outliers(df_santander['age'])
outliers_renta = detect_outliers(df_santander['renta'])

# Mostrar resultados
print("\nOutliers en 'age':")
print(outliers_age)
print("\nOutliers en 'renta':")
print(outliers_renta)
```

```python
# 17.	Calcular el intervalo de confianza para el promedio de renta al 95%. Decida si usa una Normal o una T-Student.

#Cargar librería scipy.sats importar t
from scipy.stats import t


# Calcular el intervalo de confianza
renta_mean = df_santander['renta'].mean()
renta_std = df_santander['renta'].std()
n = len(df_santander['renta'].dropna())
sem = renta_std / np.sqrt(n)  # Error estándar de la media
conf_interval = t.interval(0.95, df=n-1, loc=renta_mean, scale=sem)

# Imprimir el intervalo de confianza con formato legible
print(f"\nIntervalo de confianza para el promedio de renta (95%):")
print(f"Límite inferior: {conf_interval[0]:,.2f}")
print(f"Límite superior: {conf_interval[1]:,.2f}")
```
